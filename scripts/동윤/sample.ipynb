```python
import requests
import os
import dotenv
import pandas as pd
import time

dotenv.load_dotenv()

authKey = os.getenv('PUBLIC_DATA_API_KEY')
year = '2018074' # 보통 최신 확정 데이터인 2023년이나 2022년을 사용하세요
sido = '28'   # 인천광역시 코드

# 1. 인천광역시의 시군구 코드 리스트 (중구, 동구, 미추홀구, 연수구, 남동구, 부평구, 계양구, 서구, 강화군, 옹진군)
gu_codes = ['110', '140', '177', '185', '200', '237', '245', '260', '710', '720']

all_data = [] # 각 구별 데이터를 담을 리스트

for gu in gu_codes:
    url = f'https://opendata.koroad.or.kr/data/rest/frequentzone/lg?authKey={authKey}&searchYearCd={year}&sido={sido}&guGun={gu}&type=json&numOfRows=100'
    
    response = requests.get(url)
    data = response.json()
    
    # 데이터가 정상적으로 들어있는지 확인 ('items'와 'item'이 있는지)
    if data.get('items') and data['items'].get('item'):
        items = data['items']['item']
        all_data.extend(items) # 리스트에 데이터 추가
        print(f"✅ 구 코드 {gu}: {len(items)}건 로드 완료")
    else:
        print(f"❌ 구 코드 {gu}: 데이터 없음")
    time.sleep(0.2)

if all_data:
    incheon_all_df = pd.DataFrame(all_data)
    print("-" * 50)
    print(f"✨ 인천광역시 전체 데이터 총 {len(incheon_all_df)}건 수집 완료!")
else:
    print("수집된 데이터가 없습니다.")

# 3. 결과 출력 (판다스 표 형태)
incheon_all_df.head()
```

```python
## openai 동작 안될 시 구현 ##
import pandas as pd

df = pd.read_csv('17_24_lg.csv', encoding='cp949') # 한글 깨짐 방지를 위해 cp949 사용

# 2. 인천광역시 데이터만 추출
# '시도시군구명' 컬럼에서 '인천광역시'가 포함된 행만 필터링합니다.
incheon_df = df[df['시도시군구명'].str.contains('인천', na=False)].copy()

# 3. 데이터 기본 정보 확인
print(f"전체 데이터 수: {len(df)}건")
print(f"인천광역시 데이터 수: {len(incheon_df)}건")
print("-" * 50)

# 4. 인천에서 사고건수가 가장 많은 상위 5개 지점 확인
top_spots = incheon_df[['시도시군구명', '지점명', '사고건수', '사망자수', '중상자수']].sort_values(by='사고건수', ascending=False).head(5)

print("[인천광역시 사고 다발 지점 Top 5]")
print(top_spots)

# 5. 수치 데이터 기초 통계 (사고건수, 사상자수 등)
print("-" * 50)
print("[인천 사고 데이터 요약]")
print(incheon_df[['사고건수', '사상자수', '사망자수', '중상자수', '경상자수']].describe())
```

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv('17_24_lg.csv', encoding='cp949')

# 2. 인천광역시 데이터 필터링
incheon_df = df[df['시도시군구명'].str.contains('인천', na=False)].copy()

# 3. 요약 정보 출력 (간단하게)
print(f"✅ 전체 데이터: {len(df):,}건 | 인천광역시 데이터: {len(incheon_df):,}건")
print("-" * 80)

# 4. 인천 사고 다발 지점 Top 5 (표 형식)
# .reset_index(drop=True)를 추가하면 순위(0~4)가 깔끔하게 보입니다.
top_spots = incheon_df[['시도시군구명', '지점명', '사고건수', '사상자수', '사망자수', '중상자수']] \
            .sort_values(by='사고건수', ascending=False) \
            .head(10) \
            .reset_index(drop=True)

print("📍 [인천광역시 사고 다발 지점 Top 10]")
display(top_spots)  # 표 형태로 출력

# 5. 기초 통계 요약 (표 형식)
stats_summary = incheon_df[['사고건수', '사상자수', '사망자수', '중상자수', '경상자수']].describe()

print("\n📊 [인천 사고 데이터 통계 요약]")
display(stats_summary)
```
