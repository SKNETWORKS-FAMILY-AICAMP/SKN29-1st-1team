```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
load_dotenv()

service_key = os.getenv('PUBLIC_DATA_API_KEY')
url = 'http://apis.data.go.kr/6280000/ICRoadVolStat/NodeLink_Trfc_DD'

params = {
    'serviceKey' : service_key,
    'pageNo' : 1,
    'numOfRows' : 4,
    'YMD': 20240101,
    '_type' : 'json'
}
response = requests.get(url,params=params)
# print(f'코드 : {response.status_code}')
# print(response.json()) 

root = response.json()

header = root['response']['header']
body = root['response']['body']

items = body['items']

if isinstance(items, dict):
    items = [items]

df = [['statDate','roadName','linkID']]
df = pd.DataFrame(items)

df = df.rename(columns={
    'roadName': '도로명',
    'trafficVolume': '교통량',
    'ymd': '날짜'
})
df.to_csv('traffic_data.csv', index=False, encoding='utf-8-sig')
print(df.head())
```
