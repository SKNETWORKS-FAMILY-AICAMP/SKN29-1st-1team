```python
import requests
import os
import dotenv

dotenv.load_dotenv()

url = 'http://apis.data.go.kr/6280000/ICRoadVolStat/NodeLink_Trfc_DD'
SERVICE_KEY = os.getenv('PUBLIC_DATA_API_KEY')

params = {
    "serviceKey": SERVICE_KEY,
    "pageNo": 1,
    "numOfRows": 10,
    "YMD": 20230403
}

res = requests.get(url, params=params)

data = res.json()
items = data['response']['body']['items']

print(type(items))
```

```python
hours = [f'hour{str(i).zfill(2)}' for i in range(24)]

for item in items:
    total_traffic = sum(int(item[h]) for h in hours)
    item['total_traffic'] = total_traffic

print(items)

```

```python
import pandas as pd

df = pd.DataFrame(items)

hours = [f'hour{str(i).zfill(2)}' for i in range(24)]

df[hours] = df[hours].astype(int)
df['total_traffic'] = df['total_traffic'].astype(int)

df['peak_traffic'] = df[hours].max(axis=1)

df['stress_score'] = (
    df['total_traffic'] * 0.5 +
    df['peak_traffic'] * 0.5
)

df_sorted = df.sort_values(by='stress_score', ascending=False)

print(df_sorted[['roadName', 'startName', 'endName', 'stress_score']].head(10))
```
